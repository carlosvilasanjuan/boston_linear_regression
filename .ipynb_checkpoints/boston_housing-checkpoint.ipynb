# Methods:

__Pearson product moment correlation__
- The __Pearson correlation evaluates the linear relationship between two continuous variables(Strength & Direction)__. A relationship is linear when a change in one variable is associated with a proportional change in the other variable.


- For example, you might use a Pearson correlation to evaluate whether increases in temperature at your production facility are associated with decreasing thickness of your chocolate coating.

- __Assumptions__:
  - Data is in interval or ratio
  - No significan't outliers
  - Normal distribution of the data
 
__Spearman rank-order correlation__
__The Spearman correlation evaluates the monotonic relationship between two continuous or ordinal variables__. 
- In a __monotonic relationship, the variables tend to change together, but not necessarily at a constant rate__. The Spearman correlation coefficient is based on the ranked values for each variable rather than the raw data.

- Spearman doesen't assume normality of the data

- Spearman correlation is often used to evaluate relationships involving ordinal variables. For example, you might use a Spearman correlation to evaluate whether the order in which employees complete a test exercise is related to the number of months they have been employed.

__Kendall rank correlation:__
Kendall rank correlation is a __non-parametric test__ that __measures the strength of dependence between two variables.

Model 1

```python


x = pd.DataFrame(df['LSTAT'])
y = pd.DataFrame(df['MEDV'])

X_train_1, X_test_1, Y_train_1, Y_test_1 = train_test_split(x, y, test_size = 0.2)
```

```python
# Training and Testing throught Linear Regression Model

#Creating and Fitting LRM
lr = LinearRegression(fit_intercept=True)
lr.fit(X_train_1, Y_train_1)
```

```python

y_train_pred = lr.predict(X_train_1)
rmse = (np.sqrt(mean_squared_error(Y_train_1, y_train_pred)))

r2 = round(lr.score(X_train_1, Y_train_1),2)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```

## Model 2:
- Standarizing and shuffling the data

```python
df2 = shuffle(df)

x = pd.DataFrame(df2['LSTAT'])
y = pd.DataFrame(df2['MEDV'])


x_standard = preprocessing.scale(x)
y_standard = preprocessing.scale(y)

X_train_2, X_test_2, Y_train_2, Y_test_2 = train_test_split(x_standard, y_standard, test_size = 0.2)

lr = LinearRegression(fit_intercept=True)
lr.fit(X_train_2, Y_train_2)


y_train_pred = lr.predict(X_train_2)
rmse = (np.sqrt(mean_squared_error(Y_train_2, y_train_pred)))

r2 = round(lr.score(X_train_2, Y_train_2),2)

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")
```
